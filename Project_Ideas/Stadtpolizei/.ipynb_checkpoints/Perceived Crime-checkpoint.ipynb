```python
def get_pressreleases(row):
    data = list()
    for item in row:
        press_dict = {
            'date': 0,
            'title': 0,
            'lead': 0,
            'url': 0,
            'text': 0,
            'keys': 0
            }
        press_dict['date'] = item.find('time').text #day
        press_dict['title'] = item.find('h3').text
        press_dict['lead'] = item.find('p').text #lead
        press_dict['url'] = item.find('a').attrs['href'] #url
        data.append(press_dict)
    
    return data
```

```python
base_url = "https://www.stadt-zuerich.ch"
data = list()
for year in ["2016","2017","2018"]:
    raw_html = requests.get(base_url + "/pd/de/index/stadtpolizei_zuerich/medien/medienmitteilungen/"+year+".html")
    soup_doc = BeautifulSoup(raw_html.content, "html.parser")
    preleases = soup_doc.find_all(class_='masonrygrid_item')
    data.append(get_pressreleases(preleases))
data
```

```python
#got a list in a list. 0 = 2016, 1 = 2017, 2 = 2018
data_text = list()
for year in range(0, len(data)):
    for release in data[year]:
        raw_html = requests.get(base_url + release['url'])
        print(release['url'])
        soup_doc = BeautifulSoup(raw_html.content, "html.parser")
        text = soup_doc.find('div', attrs={'id': 'pressrelease_text'}).text.replace("\n", " ")
        text = text.replace("\xa0", " ")
        data_text.append(release)
```

```python
df = pd.DataFrame(data_text)
df = df[['date', 'title', 'lead', 'url', 'keys', 'text']]
df.to_csv("medienmitteilungen.csv", index=False)
```
