```python
import pandas as pd
import requests
import time 
import datetime
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from cycler import cycler
plt.rcParams.update(
    {'pdf.fonttype':42,
    'ps.fonttype': 42
})
import numpy
import re
import seaborn as sns

```

```python
df_gemeinde = pd.read_csv('urbanrural.csv', delimiter=";", encoding="Latin 1")
df_gemeinde.head()
```

```python
df_gemeinde = df_gemeinde[df_gemeinde['Ja in %']!='...']
df_gemeinde = df_gemeinde[df_gemeinde['Ja in %']!="nan"] 
len(df_gemeinde)
```

```python
df_newcol = pd.DataFrame(df_gemeinde['Datum und Vorlage'].str.split(' ',1).tolist(),
                                   columns = ['date','Vorlage'])
df_gemeinde[['date', 'Vorlage']] = df_newcol
#df_schweiz.columns = ['Gemeinde', 'Date_Vorlage', 'Ja', 'date', 'Vorlage']
df_gemeinde['majority']= df_gemeinde['Ja in %'].astype(float)>=50
df_gemeinde.head()
```

```python
df_schweiz = pd.read_csv('schweiz.csv', delimiter=";", encoding="Latin 1", index_col=None)
df_schweiz.head()
```

```python
df_newcol = pd.DataFrame(df_schweiz['Datum und Vorlage'].str.split(' ',1).tolist(),
                                   columns = ['date','Vorlage'])
df_schweiz[['date', 'Vorlage']] = df_newcol
#df_schweiz.columns = ['Gemeinde', 'Date_Vorlage', 'Ja', 'date', 'Vorlage']
df_schweiz['majority']= df_schweiz['Ja in %']>=50
df_schweiz.head()
```

```python
df_schweiz.majority.replace([False, True], ["1", "0"], inplace=True)
df_gemeinde.majority.replace([False, True], ["1", "0"], inplace=True)
```

```python
df = pd.merge(df_gemeinde[['Datum und Vorlage','Gemeinde', 'date', 'Vorlage', "Ja in %", 'majority']], df_schweiz[['Datum und Vorlage', 'majority', "Ja in %"]], on='Datum und Vorlage')
df.head()
```

```python
df.columns = ['Datum und Vorlage','county', 'date', 'title', "yes_county", 'majority_county', 'majority_ch','yes_ch']
df.head()
```

```python
df['comparison']= df['majority_county']==df['majority_ch']
df['year'] = df['Datum und Vorlage'].str.extract('(\d\d\d\d)')
```

```python
df=df[df['year'].astype(int)>=2000]
df=df[(df['year'].astype(int)>=2000) & (df['yes_ch'].astype(int)>=40) & (df['yes_ch'].astype(int)<=60)]
```

```python
df[df['county'].str.strip().astype(str)=='Zürich'].sort_values('year', ascending=False)
```

```python
df_count =df.groupby('county').comparison.value_counts().unstack().reset_index()
df_count.head()
```

```python
df_count['sum'] = df_count[False]+df_count[True]
df_count['ratio'] = df_count[True]/df_count['sum']*100
df_count.head()
```

```python
df_count['county']=  df_count['county'].str.strip().astype(str)
df_count.sort_values('ratio',ascending=True).head(5)
```

```python
best = df_count.sort_values('ratio',ascending=False).head(5).county.values.tolist()
best
```
