```python
import requests
import pandas as pd
import time 
import datetime
```

# Connect to members of Gemeinderat API

```python
url = "http://www.gemeinderat-zuerich.ch/api/Mitglieder/"
response = requests.get(url)
data = response.json()
```

## change it into a df

```python
df = pd.DataFrame(data, columns=data[0].keys())
df.head()
```

Search for motions by politician
http://www.gemeinderat-zuerich.ch/api/Geschaeft/suchen?personId=afb2e0bc-b941-4fa4-8936-9b4978fc1ee1
  
  there are different pages to get the results &activePage=1

```python
url_geschaeft = "http://www.gemeinderat-zuerich.ch/api/Geschaeft/suchen?personId="

politicians_list = [] #for bringing all together in a list of dict.
counter = int(-1) #to count the politicians number. is needed to get string elements out of the objects.
pages = list(range(1, 21)) #random set number of maximal pages run. highest number * 10 should exceed politicans max motiions

#loop for every politician ID
for politician in df.Id:
    print("----")
    
    #making an empty dict to fill
    politicians = {}
    
    #filling up the counter per politician
    counter = counter + 1
    
    #start with the different pages for each politician
    for page in pages:

        response = requests.get(url_geschaeft+politician+"&activePage="+str(page))
        data_polit = response.json()
        
        #if the are no motions on a page, brake the loop and go to the next politician
        if len(data_polit['Geschaefte']) == 0:
            break
            
        else:
            print(df[df.Id==politician].Name[counter]) #just for seeing the script working
            print(counter)
            #we need another loop to get all the motions
            for vorstoss in data_polit['Geschaefte']:           
            
                politicians['id'] = politician   
                politicians['Name'] = df[df.Id==politician].Name[counter]
                politicians['Vorname'] = df[df.Id==politician].Vorname[counter]
                politicians['Partei'] = df[df.Id==politician].Partei[counter]
                politicians['Wahlkreis'] = df[df.Id==politician].Wahlkreis[counter] 
                politicians['Anzahl_Vorstoesse'] = data_polit['AnzahlResultate']
                politicians = {**politicians, **vorstoss} #merging two lists together
                politicians['time_scraped'] = datetime.datetime.fromtimestamp(time.time()).strftime('%Y-%m-%d %H:%M:%S')
                politicians_list.append(politicians) #append all of it to one list of dict
```

```python
df_vorstoesse = pd.DataFrame(politicians_list, columns=politicians_list[0].keys())
df_vorstoesse
```
