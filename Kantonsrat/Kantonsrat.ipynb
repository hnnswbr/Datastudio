```python
#import the AU-PDF
pdfFileObj = open('Protokolle/129_2015-2019.pdf', 'rb')
pdfFileObj
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
pageObj = pdfReader.getPage(2)
protokoll = pageObj.extractText()
protokoll = protokoll.replace("\n", "").strip()
```

```python
data = list()
text = str()
for page in list(range(0,pdfReader.numPages-1)):
    pageObj = pdfReader.getPage(page)
    protokoll = pageObj.extractText()
    #print(page)
    protokoll = protokoll.replace("\n", "").strip()
    text = text + " pagebreack " + protokoll

```

```python
text
```

```python
politicians = re.split(r"([\w-]* \w* [(]\w*, [\w\. -]*[)])[:,]*[\w\(\)\s]*:",text)
```

```python
data = list()
for index in range(1,len(politicians)):
    places = {
        'politiker': 0,
        'text': 0
        }
    if index%2!=0:
        politiker = politicians[index]
    else:
        speechtext = politicians[index]
        speeches = {
            'politiker': politiker,
            'text': speechtext
        }
        data.append(speeches)
```
